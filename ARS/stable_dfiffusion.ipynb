```python
device = "cuda"
model_id_or_path = "runwayml/stable-diffusion-v1-5"
model_path = "./output/pytorch_lora_weights.bin"

pipe = StableDiffusionImg2ImgPipeline.from_pretrained(model_id_or_path, 
                                                      #torch_dtype=torch.float16,   #cuda 사용시에만 사요
                                                      safety_checker = None)
pipe = pipe.to(device)

pipe.unet.load_attn_procs(model_path)
# set to use GPU for inference
pipe.to(device)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from tqdm.notebook import tqdm_notebook

# 이미지 크기 조정 함수 정의
def resize_image(image, size):
    return image.resize(size)

diffusion_imgs = []
prompt = "standing"

strengths = np.arange(0.15, 0.95, 0.10)
guidance_scale = np.arange(1, 4)
multi_index = pd.MultiIndex.from_product([strengths, guidance_scale], names=['Strength', 'Guidance Scale'])

# 이미지 불러오기 및 크기 조정
image = Image.open('3.jpg')
image = resize_image(image, (244, 244))

fig, axes = plt.subplots(len(strengths) + 1, len(guidance_scale), figsize=(10, 14))

for idx in range(0, len(guidance_scale)):
    axes[0, idx].axis('off')
    axes[0, idx].imshow(image)
    axes[0, idx].set_title('input image', fontsize=8)

for i, (strs, guid) in tqdm_notebook(enumerate(multi_index)):
    images = pipe(prompt=prompt, image=image, guidance_scale=guid, strength=strs).images
    diffusion_imgs.append(images[0])
    ax = axes[(i + len(guidance_scale)) // len(guidance_scale), (i + len(guidance_scale)) % len(guidance_scale)]
    ax.imshow(images[0])
    ax.set_title(f'Strength: {strs:.2f}, Guidance: {guid}', fontsize=8)
    ax.set_xticks([])
    ax.set_yticks([])
    plt.imsave(f"./data/trail1/diffusion_img_guidance_scale={guid}_strength={strs}.png", np.array(images[0]))


fig.suptitle("Images with Different Strengths and Guidance Scales", fontsize=24, y=1.01)
fig.subplots_adjust(wspace=0.1, hspace=0.1)
fig.tight_layout(pad=0.0)
plt.show()
```
